```python
def userMel(fileName):
    myFile = open(fileName,"r")
    myFile.readline()
    stats=myFile.read().splitlines()
    myFile.close()
    
    totalWin = 0
    totalLoss = 0
    totalGames = 10
    totalHealing = 0    
    totalRes = 0
    totalDeath = 0
    cardCount = 0
    totalKp = 0
    kpCount = 0
    totalHp = 0
    hpCount = 0
    totalOfA = 0
    totalDfA = 0
    totalKills = 0 
    totalDmg = 0
    kpList = []
    hpList = []
    statList = []
    
    for data in stats:
        statParts = data.split(",")
        winLoss = statParts[6]
        user = statParts[2]
        if((user== "melona") and (winLoss== ("win"))):
            totalWin += 1
        elif((user== "melona") and (winLoss== ("loss"))):
            totalLoss += 1
    totalGames =(totalWin + totalLoss)   
    lossPercentageDecimal = (totalLoss/totalGames)
    lossPercentage = lossPercentageDecimal * 100
    winPercentageDecimal = (totalWin/totalGames)
    winPercentage = winPercentageDecimal * 100
    
    print("The number of Games played for melona is: " + str(totalGames))
    print("The overall win percentage for all games played for melona is: " + str(round(winPercentage,2))+ "%" )
    print("The overall loss percentage for all games played for melona is: " + str(round(lossPercentage,2))+ "%" )

    
    for data in stats:
        statParts = data.split(",")
        healing = int(statParts[11])
        if("melona" in statParts):
            totalHealing += healing
            avgHealing = (totalHealing/totalGames) 
            
            res = int(statParts[12])
            totalRes += res
            avgRes = (totalRes/totalGames)

            death = int(statParts[13])
            totalDeath += death
            avgDeath =(totalDeath/totalGames)
            resToDeathRatio = (avgRes/avgDeath)
            
            ofAssist = int(statParts[16])
            dfAssist = int(statParts[17])
            totalOfA += ofAssist
            totalDfA += dfAssist
            avgOfA = (totalOfA/totalGames)
            avgDfA = (totalDfA/totalGames)
            
            kills = int(statParts[18])
            totalKills += kills
            avgKills = (totalKills/totalGames)
            
            dmgBoost = int(statParts[19])
            totalDmg += dmgBoost
            avgDmg =(totalDmg/totalGames)
            
            killingPartCard = statParts[14]
            healingPercCard = statParts[15]

            if(len(killingPartCard) != 0):
                kpList.append(killingPartCard)
                cardCount +=1
                for num in range (len(kpList)):
                    kpList[num]= int(kpList[num])                    
            if(len(healingPercCard) != 0):
                hpList.append(healingPercCard)
                cardCount +=1
                for num in range (len(hpList)):
                    hpList[num]=int(hpList[num])

        for num in kpList:
            totalKp += num
            kpCount += 1

        for num in hpList:
            totalHp += num  
            hpCount += 1
   
    avgCards = (cardCount/totalGames) * 100
    avgKp = (totalKp/kpCount) 
    avgHp= (totalHp/hpCount)            
    
    print("The total healing done for all games played for melona is: " + str(totalHealing))
    print("The average healing done per game played for melona is: " + str(round(avgHealing)))
    
    print("The number of successful resurrections of teammates for all games played for melona is: " + str(totalRes))
    print("The average for successful resurrections of all teammates per game played for melona is: " + str(round(avgRes,2)))
    print("The average of deaths per game played for melona is: " + str(round(avgDeath,2)))
    print("The average ressurrection to death ratio per game played for melona is: " + str(round(resToDeathRatio,2)))
    
    print("The average number of offensive assists per game for melona is: " + str(round(avgOfA)))
    print("The average number of deffensive assists per game for melona is: " + str(round(avgDfA)))
    
    print("The average number of blaster kills per game for melona is: " + str(round(avgKills))) 
    
    print("The average number of damage boosted per game for melona is: " + str(round(avgDmg)))
    
    print("The number of overall cards won for having best statistics in a game for melona is: "+ str(cardCount))
    print("The chance of winning a best statistics card per game for melona is: " + str(round(avgCards,2)) + "%")
    print("The average kill participation per game if a card is won for melona is: " + str(round(avgKp,2)) + "%")
    print("The average healing participation per game if a card is won for melona is: " + str(round(avgHp,2)) + "%")
```
